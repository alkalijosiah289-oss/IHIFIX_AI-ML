```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/diabetes.csv")
```

```python
df.groupby("Outcome")["BloodPressure"].mean().plot(kind="bar")

plt.title("Average Blood Pressure by Diabetes Outcome")
plt.xlabel("Outcome")
plt.ylabel("Average Blood Pressure")
plt.show()
```

Patients with diabetes had an average blood pressure of 70.82 compared to 68.18 for non-diabetic patients.

```python
df.groupby("Outcome")["Insulin"].mean().plot(kind="bar")

plt.title("Average Insulin by Diabetes Outcome")
plt.xlabel("Outcome")
plt.ylabel("Average Insulin Level")
plt.show()
```

Patients with diabetes had an average insulin level of 100.34 compared to 68.79 for non-diabetic patients.

This suggests patients with diabetes generally had higher insulin measurements.

```python
df.groupby("Pregnancies")["Outcome"].mean().plot(kind="line")

plt.title("Diabetes Rate by Number of Pregnancies")
plt.xlabel("Number of Pregnancies")
plt.ylabel("Diabetes Rate")
plt.show()
```

Patients with 9 pregnancies had a diabetes rate of 64.3%, while patients with 1 pregnancy had a diabetes rate of 21.5%.
