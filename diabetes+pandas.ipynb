```python
import pandas as pd

df = pd.read_csv("/content/diabetes.csv")
diabetes_count = (df["Outcome"] == 1).sum()
print("Patients with diabetes:", diabetes_count)
```

```python
import pandas as pd

df = pd.read_csv("/content/diabetes.csv")
avg_glucose = df["Glucose"].mean()
print("Average glucose level:", avg_glucose)
```

```python
import pandas as pd

df = pd.read_csv("/content/diabetes.csv")
patients_above_40 = df[df["Age"] > 40]
print(patients_above_40)
```

```python
import pandas as pd

df = pd.read_csv("/content/diabetes.csv")
avg_bmi = df.groupby("Outcome")["BMI"].mean()
print(avg_bmi)
```

```python
import pandas as pd

df = pd.read_csv("/content/diabetes.csv")
rows_removed = len(df) - len(df[(df["Glucose"].notna()) & (df["Glucose"] != 0)])
print("Rows removed:", rows_removed)
```
