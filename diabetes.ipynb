```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/diabetes.csv")
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Glucose"])
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Number of Patients")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["Glucose"])
plt.title("Age vs Glucose")
plt.xlabel("Age")
plt.ylabel("Glucose Level")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["Outcome"].value_counts().plot(kind="bar")
plt.title("Diabetes Outcome Count")
plt.xlabel("Outcome")
plt.ylabel("Number of Patients")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(df["BMI"])
plt.title("BMI Distribution")
plt.xlabel("Patients")
plt.ylabel("BMI")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Glucose"])
plt.title("Glucose Levels Across Records")
plt.xlabel("Patient Index")
plt.ylabel("Glucose Level")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Glucose"])
plt.title("Distribution of Glucose Levels")
plt.xlabel("Glucose Level")
plt.ylabel("Number of Patients")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Age"], y=df["Glucose"])
plt.title("Age vs Glucose")
plt.xlabel("Age")
plt.ylabel("Glucose Level")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df["Outcome"])
plt.title("Diabetes Outcome Count")
plt.xlabel("Outcome")
plt.ylabel("Number of Patients")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(y=df["BMI"])
plt.title("BMI Distribution")
plt.xlabel("Patients")
plt.ylabel("BMI")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=df.index, y=df["Glucose"])
plt.title("Glucose Levels Across Records")
plt.xlabel("Patient Index")
plt.ylabel("Glucose Level")
plt.show()
```

Preferred style: Seaborn-styled charts

Why: Seaborn creates cleaner and more visually appealing charts with less code and better default formatting, making the data easier to understand.
